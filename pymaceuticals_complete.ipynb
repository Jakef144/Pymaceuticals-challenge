### Analysis

- The summary statistics and box plots show that Capomulin and Ramicane have lower average tumor volumes compared to other treatments. This suggests that these two drug regimens are more effective in reducing tumor size.
  
- The scatter plot and linear regression analysis for the Capomulin regimen show a positive correlation between mouse weight and tumor volume. This suggests that heavier mice tend to have larger tumor volumes, although the correlation is moderate, indicated by the correlation coefficient.
This correlation could imply that mouse weight is a factor in tumor development or that the tumor size affects the overall health and weight of the mice.

- The line plot for a single mouse treated with Capomulin shows a clear downward trend in tumor volume over time. This indicates that the Capomulin regimen is effective at steadily reducing tumor size as treatment progresses.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df = pd.merge(study_results, mouse_metadata, on='Mouse ID') # On command looks for matching values in mouse id column

# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
number_of_mice = combined_df['Mouse ID'].nunique()
number_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = combined_df[combined_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)] #Look for the data by comparing the two fields
duplicate_mice_ids = duplicate_mice['Mouse ID'].unique() # Pull the unique data
duplicate_mice_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouse_data = combined_df[combined_df['Mouse ID'] == 'g989']
# Displaying the data
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
rows_to_drop = combined_df[combined_df['Mouse ID'] == 'g989'].index
cleaned_df = combined_df.drop(rows_to_drop)
cleaned_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
number_of_mice_cleaned = cleaned_df['Mouse ID'].nunique()
number_of_mice_cleaned
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_statistics = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    Mean_Tumor_Volume='mean',
    Median_Tumor_Volume='median',
    Tumor_Volume_Variance=lambda x: x.var(),
    std_dev='std',
    SEM=lambda x: x.sem()
)
summary_statistics
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_advanced = cleaned_df.groupby('Drug Regimen').agg(
    Mean_Tumor_Volume=('Tumor Volume (mm3)', 'mean'),
    Median_Tumor_Volume=('Tumor Volume (mm3)', 'median'),
    Variance_Tumor_Volume=('Tumor Volume (mm3)', 'var'),
    Std_Dev=('Tumor Volume (mm3)', 'std'),
    SEM=('Tumor Volume (mm3)', 'sem')
)

# Display the advanced summary statistics table
summary_advanced
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Filtering the data to include only the four specified treatment regimens
selected_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
filtered_df = cleaned_df[cleaned_df['Drug Regimen'].isin(selected_regimens)]

# Finding the final (maximum) timepoint for each mouse
final_timepoints = filtered_df.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merging the final timepoints with the original filtered data to get the tumor volume at the last timepoint
final_tumor_volumes = pd.merge(final_timepoints, filtered_df, on=['Mouse ID', 'Timepoint'], how='left')

# Display the final tumor volumes for each mouse in the selected regimens
final_tumor_volumes.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []


# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treatments:
    treatment_volumes = final_tumor_volumes[final_tumor_volumes['Drug Regimen'] == treatment]['Tumor Volume (mm3)']

    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_data.append(treatment_volumes)
    
    # Calculate the quartiles and IQR
    quartiles = treatment_volumes.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq


    # add subset


    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = treatment_volumes[(treatment_volumes < lower_bound) | (treatment_volumes > upper_bound)]

    # Display the results for each treatment
    print(f"{treatment}'s potential outliers: {outliers}")



```

```python
# Creating the figure
plt.figure(figsize=(10, 6))

# Plot the tumor volume data for each treatment group using a box plot
flierprops = dict(marker='o', markerfacecolor='red', markersize=10)  # Adjusting the outlier point size and color

# Create the box plot with the customized outliers
plt.boxplot(tumor_vol_data, labels=treatments, flierprops=flierprops)

# Add labels and title to the plot
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.title('Tumor Volume Distribution for Each Treatment Regimen')

# Display the plot
plt.show()
```

## Line and Scatter Plots

```python
# Filter the data for a single mouse treated with Capomulin
mouse_id = 'b128'  # Example mouse ID, replace with a valid one if needed
mouse_data = cleaned_df[(cleaned_df['Mouse ID'] == mouse_id) & (cleaned_df['Drug Regimen'] == 'Capomulin')]
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
plt.figure(figsize=(8, 6))
plt.plot(mouse_data['Timepoint'], mouse_data['Tumor Volume (mm3)'], marker='o')

# Add labels and title
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Tumor Volume vs. Time for Mouse {mouse_id} Treated with Capomulin')

# Display the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Filter the data for the Capomulin regimen
capomulin_data = cleaned_df[cleaned_df['Drug Regimen'] == 'Capomulin']

# Calculate the average tumor volume for each mouse
avg_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Get the corresponding mouse weights
mouse_weights = capomulin_data[['Mouse ID', 'Weight (g)']].drop_duplicates().set_index('Mouse ID')

# Merge the average tumor volume and mouse weights into one DataFrame
capomulin_scatter_data = pd.merge(avg_tumor_volume, mouse_weights, on='Mouse ID')

# Generate a scatter plot of mouse weight vs. average tumor volume
plt.figure(figsize=(8, 6))
plt.scatter(capomulin_scatter_data['Weight (g)'], capomulin_scatter_data['Tumor Volume (mm3)'], marker='o')

# Add labels and title
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')

# Display the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
from scipy.stats import linregress

# Extract the necessary data for correlation and linear regression
x_values = capomulin_scatter_data['Weight (g)']
y_values = capomulin_scatter_data['Tumor Volume (mm3)']

# Calculate the correlation coefficient
correlation = x_values.corr(y_values)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x_values, y_values)

# Print the results
print(f"Correlation coefficient: {correlation}")
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")

# Generate the linear regression line
regress_values = slope * x_values + intercept

# Plot the scatter plot and the linear regression line
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, marker='o')
plt.plot(x_values, regress_values, color='red')

# Add labels and title
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')

# Display the plot
plt.show()
```
